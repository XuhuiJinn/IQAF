```python
df_etf=pd.read_csv("/Users/judy/IQAF/111.csv").dropna()
df_etf["DATE"]=pd.to_datetime(df_etf["DATE"])
df_sc=pd.read_csv("/Users/judy/IQAF/slope_curve.csv").dropna()
df_sc["DATE"]=pd.to_datetime(df_sc["DATE"])
df_effr=pd.read_csv("/Users/judy/IQAF/effr.csv").dropna()
df_effr["DATE"]=pd.to_datetime(df_effr["DATE"])
df_effr.rename(columns={'PX_LAST':'EFFR'}, inplace=True)
df_l=pd.read_csv("/Users/judy/IQAF/leverage(spx).csv").dropna()
df_l["DATE"]=pd.to_datetime(df_l["DATE"])
df_fcf=pd.read_csv("/Users/judy/IQAF/SPX FCF.csv").dropna()
df_fcf["DATE"]=pd.to_datetime(df_fcf["DATE"])
df_fc=pd.read_csv("/Users/judy/IQAF/financial conditions.csv").dropna()
df_fc["DATE"]=pd.to_datetime(df_fc["DATE"])
df_cpi=pd.read_csv("/Users/judy/IQAF/CPI.csv").dropna()
df_cpi["DATE"]=pd.to_datetime(df_cpi["DATE"])
df_daily=pd.merge(df_etf,df_sc,on='DATE')
df_daily=pd.merge(df_daily,df_effr,on='DATE')
df_daily=pd.merge(df_daily,df_l,on='DATE')
df_daily=pd.merge(df_daily,df_fcf,on='DATE')
df_daily=pd.merge(df_daily,df_fc,on='DATE')
df_daily=pd.merge(df_daily,df_cpi,on='DATE')
df_daily[['S&P 500','QTEC','PSR','IYZ','IHI','KIE','PSJ','VAW','VFH','KBE','IYH']]=df_daily[['S&P 500','QTEC','PSR','IYZ','IHI','KIE','PSJ','VAW','VFH','KBE','IYH']].pct_change(1)
df_daily.shift(periods=1)
df_daily.to_csv('/Users/judy/IQAF/daily.csv')
```
